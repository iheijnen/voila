##### Identification and Reporting and Resolving of missing data and inconsistencies (#1)
In the graph above there are outliers for the years 2000, 2002, 2004, 2005, 2008, 2010, 2011, 2013 and 2015. A table was made to visualise these values.

```python
tablepopulation_thailandrq1 = rq1_thailand_df.groupby('year')['population'].sum().reset_index()
print(tablepopulation_thailandrq1)
```

In this table, it became clear that the most probable issues with the values were missplaced decimals.

```python
#seperate population df was made with only population
rq1_thailand_pop_df = rq1_thailand_df[['population']].copy()

#the misplaced decimals were corrected
rq1_thailand_pop_df['population'] = rq1_thailand_pop_df['population'].astype(str) #converted to string so it can be manipulated
rq1_thailand_pop_df['population'] = rq1_thailand_pop_df['population'].str.replace('.', '') #non-numeric characters were removed
rq1_thailand_pop_df['population'] = rq1_thailand_pop_df['population'].apply(lambda x: x[:-1] + '.' + x[-1] if len(x) > 6 else x) #decimal places were adjusted if needed
rq1_thailand_pop_df['population'] = rq1_thailand_pop_df['population'].astype(float) #converted back to float

#checked if corrected correclty by making a table
print(rq1_thailand_pop_df.head(16))
```

That approach did not work for some reason. Next, another approach was tried. The population of Thailand was checked online, and was verified that the population of Thailand was around 62 million in 2000. Setting this as a treshold, the outliers can be filtered. Then for these outliers, the decimal can be moved to the right.

```python
population_threshold = 62000000

while rq1_thailand_pop_df['population'].min() < population_threshold:

   outliers_population = rq1_thailand_pop_df['population'] < population_threshold

   rq1_thailand_pop_df.loc[outliers_population, 'population'] *= 10

print(rq1_thailand_pop_df)
```

This approach worked. The corrected population values were put back in the 'rq1_thailand_df' to replace the values with the outliers.
To check if it worked correctly, the population was plotted over time.

##### Identification and Reporting and Resolving of missing data and inconsistencies (#**3**)

The data seems to not make sense. The percentage expenditure is supposed to be a percentage of the GDP per capita spent on health care. However here, the values are mostly above 100%. A table was made to visualise.

```python
tableexp_thailandrq1 = rq1_thailand_df.groupby('year')['percentage expenditure'].sum().reset_index()
print(tableexp_thailandrq1)
```

All values, except for 2002, and 2010 do not make sense. It could be that here again, the decimal is placed on wrong. A threshold was made at 100 (%). Then, the same approach as for population was used to replace the decimal correctly.

```python
#seperate exp df was made with only expenditure percentage
rq1_thailand_exp_df = rq1_thailand_df[['percentage expenditure']].copy()
```

```python
exp_threshold = 100

while rq1_thailand_exp_df['percentage expenditure'].max() > exp_threshold:

   outliers_exp = rq1_thailand_exp_df['percentage expenditure'] > exp_threshold

   rq1_thailand_exp_df.loc[outliers_exp, 'percentage expenditure'] /= 10

print(rq1_thailand_exp_df)
```

Only the first and the last value are now still off. A linear regression method was tried to extrapolate the values in order to get more correct values for the first and last value.

```python
from sklearn.linear_model import LinearRegression

#year values extracted
years = rq1_thailand_exp_df.index.values.reshape(-1, 1)

#percentage expenditure values extracted
expenditure = rq1_thailand_exp_df['percentage expenditure'].values.reshape(-1, 1)

#linear regression model fitted
model = LinearRegression()
model.fit(years, expenditure)

#the first and last values extrapolated using the linear model
first_value = model.predict([[years[0][0] - 1]])[0][0]
last_value = model.predict([[years[-1][0] + 1]])[0][0]

#the first and last values updated in the df
rq1_thailand_exp_df.iloc[0, 0] = first_value
rq1_thailand_exp_df.iloc[-1, 0] = last_value

print(rq1_thailand_exp_df)
```

```python
rq1_thailand_df['percentage expenditure'] = rq1_thailand_exp_df['percentage expenditure']
```

```python
alt.Chart(rq1_thailand_df).mark_line().encode(
    x='year:O',
    y='percentage expenditure:Q',
    tooltip=['year', 'percentage expenditure']
).properties(
    width=600,
    height=400,
    title='Percentage expenditure Over Time (Thailand) '
).interactive()
```

Inconsistencies were solved!
